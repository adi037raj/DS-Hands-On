```python
hd=pd.read_csv('13.1 heart-disease.csv')
hd.head()
```

```python
hd.age.plot.hist(bins=10);
```

```python
hd.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50=hd[hd.age>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(x='age',y='chol',kind='scatter',c='target',figsize=(10,10))
```

```python
 # OO method with mixed pyplot

fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
#ax.set_xlim([45,90]);
```

```python
#OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);
ax.set(title='Heart disease',xlabel='Age',ylabel='Chol');
ax.legend(*scatter.legend_elements(),title='Target');
ax.axhline(y=hd.chol.mean(),linestyle='--');
```

```python
fig,(ax0,ax1)=plt.subplots(figsize=(10,10),nrows=2,sharex=True)


scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax0.set(title='Heart Disease and Cholestrol level',ylabel='Chol')
ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(y=over_50.chol.mean(),linestyle='--');



scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
ax1.set(xlabel='Age',ylabel='Thalach',title='Heart Disease and Max Heart Rate')
ax1.legend(*scatter.legend_elements(),title='Taget')
ax1.axhline(y=over_50.thalach.mean(),linestyle='--');



fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold');

```
